```python
# Se importan OS y TIME necesarias para la animación
import os
import time
import csv

# Esta función es utilizada para crear tableros, los cuales representan el ambiente de las celulas.
# Los tableros creados son de dimensión m po n. por lo que puede ser de cualquier tamaño matricial.
def tablero (m,n,z):
    """
    Función que crea tableros de dimensión m por n y los rellena de ceros, lo cual es el plano inicial
    del juego.
    """
    i = 0
    
    # A la lista creada se le agregan más listas vacías
    while i < m:
        i += 1
        z.append([])
    
    # a las listas anteriores se le agregan sublistas, lo cual nos da la representación matricial
    for j in range(m):
        for k in range(n):
            z[j].append([])
    
    # Las matrices creadas son rellenadas de ceros como condición inicial
    for j in range(m):
        for k in range(n):
            z[j][k] = 0

    return z

# Función de condiciones iniciales, la cual permite ingresar los pares ordenados de las células vivas.
def celulas_iniciales(cel,b,punto_x,punto_y):
    """
    Permite ingresar los valores de las células vivas.
    """
    # Puntos que representan las células vivas
    datos_m = []
    datos_n = []
    
    for g in range(len(punto_x)):
        datos_m.append(int(punto_x[g]))
    
    for h in range(len(punto_y)):
        datos_n.append(int(punto_y[h]))
    
    # Asigna el valor de 1
    j = 0
    while j< (cel):
        b[datos_m[j]-1][datos_n[j]-1] = 1
        j += 1
    return a

# Función de suma para esquinas
def esqu(m,n,b,c):
    """
    Es utilizada para evaluar la regla creada en el tablero, solo esquinas.
    """
    # Coordenadas de esquina
    esq_i = [0,0,-1,-1]
    esq_j = [0,-1,0,-1]
    
    # Sumatoria de los entornos
    POO = [(b[0][1]),(b[1][1]),(b[1][0])]
    P1n = [(b[0][-2]),(b[1][-2]),(b[1][-1])]
    Pm1 = [(b[-2][0]),(b[-2][1]),(b[-1][1])]
    Pmn = [(b[-2][-1]),(b[-2][-2]),(b[-1][-2])]
    esquinas = [sum(POO),sum(P1n),sum(Pm1),sum(Pmn)]
    
    # Aplica la regla para las esquinas
    for i in range(4):
        # Para las células muertas
        if b[esq_i[i]][esq_j[i]] == 0:
            if esquinas[i] == 3:
                c[esq_i[i]][esq_j[i]] = 1
            else:
                pass
        # Para las células vivas
        else:
            if 2<=esquinas[i]<4:
                c[esq_i[i]][esq_j[i]] = 1
            else:
                c[esq_i[i]][esq_j[i]] = 0

# Función de suma de cuerpo
def cuerpo(m,n,b,c):
    """
    Es utilizada para evaluar la regla creada en el tablero, para el cuerpo, no incluye contorno y esquinas.
    """
    # Evalúa la suma de los entornos
    for i in range(m-2):
        i += 1
        for j in range(n-2):
            j += 1
            total = b[i-1][j-1]+b[i-1][j]+b[i-1][j+1]+b[i][j-1]+b[i][j+1]+b[i+1][j-1]+b[i+1][j]+b[i+1][j+1]
            # Aplica la regla
            regla(b,c,i,j,total)

# Función de suma de cuerpo Izq
def cuerpo_iz(m,n,b,c):
    """
    Es utilizada para evaluar la regla creada en el tablero, contorno izquierdo.
    """
    # Evalúa la suma de los entornos
    for i in range(m-2):
        i += 1
        j = 0
        total = b[i-1][j]+b[i-1][j+1]+b[i][j+1]+b[i+1][j]+b[i+1][j+1]
        # Aplica la regla
        regla(b,c,i,j,total)

# Función de suma de cuerpo dere
def cuerpo_de(m,n,b,c):
    """
    Es utilizada para evaluar la regla creada en el tablero, contorno desecho.
    """
    # Evalúa la suma de los entornos
    for i in range(m-2):
        i += 1
        j = -1
        total = b[i-1][j-1]+b[i-1][j]+b[i][j-1]+b[i+1][j-1]+b[i+1][j]
        # Aplica la regla
        regla(b,c,i,j,total)

# Función de suma de cuerpo abajo
def cuerpo_aba(m,n,b,c):
    """
    Es utilizada para evaluar la regla creada en el tablero, contorno de abajo.
    """
    # Evalúa la suma de los entornos
    i = -1
    for j in range(n-2):
        j += 1
        total = b[i-1][j-1]+b[i-1][j]+b[i-1][j+1]+b[i][j-1]+b[i][j+1]
        # Aplica la regla
        regla(b,c,i,j,total)

# Función de suma de arriba
def cuerpo_arr(m,n,b,c):
    """
    Es utilizada para evaluar la regla creada en el tablero, contorno de arriba.
    """
    # Evalúa la suma de los entornos
    i = 0
    for j in range(n-2):
        j += 1
        total = b[i][j-1]+b[i][j+1]+b[i+1][j-1]+b[i+1][j]+b[i+1][j+1]
        # Aplica la regla
        regla(b,c,i,j,total)

# Reglas
def regla(b,c,i,j,total):
    """
    Regla de la vida:
    Una célula muerta con exactamente 3 células vecinas vivas "nace".
    Una célula viva con 2 o 3 células vecinas vivas sigue viva, en otro caso muere.
    """
    # Regla de las céluas muertas
    if b[i][j] == 0:
        if total == 3:
            c[i][j] = 1
        else:
            pass
    # Regla de las céluas vivas
    else:
        if 2<=total<4:
            c[i][j] = 1
        else:
            c[i][j] = 0

# Limpiar matriz
def copiar(m,n,b,c):
    """
    Traslada los valores de la matriz de transición a la matriz de iteración.
    """
    # Traslada la información de una matriz a otra
    for j in range(m):
        for k in range(n):
            b[j][k] = c[j][k]

# animación
def anima(c):
    """
    Permite llevar los resultados en terminos de animación.
    """
    # Lleva a la animación * = 1 y " "=0
    for j in range(m):
        for k in range(n):
            if c[j][k] == 0:
                c[j][k] = " "
            else:
                c[j][k] = "*"

# desanimación                
def desanima(c):
    """
    Regresa los resultados a su estado original.
    """
    # Lleva a la animación * = 1 y " "=0
    for j in range(m):
        for k in range(n):
            if c[j][k] == " ":
                c[j][k] = 0
            else:
                c[j][k] = 1
    
# Tiempo y población
def pobla(c):
    """
    Devuelve el momento (tiempo) y la población de las células.
    """
    población2 = []
    parcial = ""
    for i in range(len(c)):
        parcial = sum(c[i])
        # Acumula la población para cada momento
        población2.append(parcial)
    población.append(sum(población2))
    
# Extraer la información
def extrae(par):
    """
    Extrae los pares ordenados iniciales.
    """
    for p in range(len(par)):
        temporal = par[p].split(",")
        punto_x.append(temporal[0])
        punto_y.append(temporal[1])
        temporal = ""

# tabla salida
def tabla_salida():
    with open('elmer_lemus.txt', mode='w') as salida_archivo:
        salida_escritura = csv.writer(salida_archivo, delimiter=',', quotechar='"')
        salida_escritura.writerow(['Momento', 'Población'])
        for key, values in direct.items():
            salida_escritura.writerow([key, values])
```

```python
# Crear el tablero inicial
# Extrae los datos

nombre_archivo = str(input("Ingrese el nombre del archivo de instrucciones: "))
nombre_archivo = nombre_archivo+".txt"
nombre_archivo

f = open(nombre_archivo, 'r')
contenido = f.read()
f.close()

# Coloca los datos en forma de lista
datos = contenido.split("\n")
generales = datos[0].split(",")
par = datos[1:]
punto_x = []
punto_y = []

# Extrae la información y la coloca en listas
extrae(par)

m = int(generales[0])
n = int(generales[1])

# Matrices vacías
a = [] # Matriz inicial
b = [] # Matriz respuesta
c = [] # Matriz de transición

# Relleno de matrices
tablero(m,n,a)
tablero(m,n,b)
tablero(m,n,c)

# Ingreso de valores iniciales
cel = len(par)
celulas_iniciales(cel,b,punto_x,punto_y)

# os.System escribe en la terminal lo que está adentro del paréntesis. Limpia la consola.
os.system('cls')

# Imprime la matriz inicial
print("\n","Matriz inicial","\n")
print("\n","Células vivas con 1")
print("Células muertas con 0","\n")
for i in range(m):
    print(*b[i])

# Da el tiempo de espera para la impresión.
time.sleep(2)
    
# Tiempo y población
tiempo = []
población = []
    
# os.System escribe en la terminal lo que está adentro del paréntesis. Limpia la consola.
os.system('cls')

# Proceso iterativo
proce = 0
ite = int(generales[2])
while proce < ite:
    proce += 1
    print(" ")
    
    # Aplicación de las reglas
    esqu(m,n,b,c)
    cuerpo(m,n,b,c)
    cuerpo_iz(m,n,b,c)
    cuerpo_de(m,n,b,c)
    cuerpo_aba(m,n,b,c)
    cuerpo_arr(m,n,b,c)
    
    # Impresión de salida
    anima(c)
    print("\n","Matriz de transición", proce,"\n")
    for i in range(m):
        print(*c[i])
    # Da el tiempo de espera para la impresión.
    time.sleep(0.5)
    # os.System escribe en la terminal lo que está adentro del paréntesis.
    os.system('cls')
    desanima(c)
    pobla(c)
    tiempo.append(proce)
    copiar(m,n,b,c)

# Crea un directorio
direct = {}

# Llena el directorio con los valores de tiempo y población
for i in range(len(tiempo)):
    direct[tiempo[i]] = población[i]

# Guarda el archivo csv en excel o en texto blog de notas
tabla_salida()

# Fin del programa
```
